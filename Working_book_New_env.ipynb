```python
prettymaps.create_preset(
    "waterways_and_boundaries",
    layers = {
        "water": {
            "tags": {
                "natural": ["water", "bay"]
            }
        },
        "perimeter": {}
    },
    style = {
        "water": {
            "fc": "#a1e3ff",
            "ec": "#2F3737",
            "lw": 1,
            "zorder": 4
        },
        "perimeter": {
            "fill": False,
            "lw": 0,
            "zorder": 0
        },
        "text": {
            "findfont": "Century Schoolbook",
            "fontstyle": "normal",
            "fontcolor": "#000000",
            "fontsize": 12,
            "zorder": 5
        }
    }
)

# To use this new preset for a specific location
# fig, ax = prettymaps.plot('Lansing, Michigan, USA', preset='waterways_and_boundaries', radius=False)

plot = prettymaps.plot('Lansing, Michigan, USA', preset='waterways_and_boundaries', radius=False)
# plot.show()

```

```python
# 2. A map showing only large roads:

# We'll assume "large roads" to mean motorways, trunk roads, and primary roads:

plot_large_roads = prettymaps.plot(
    'Lansing, Michigan, USA',
    layers = {
        "streets": {
            "tags": {
                "highway": ["motorway", "trunk", "primary"]
            }
        }
    },
    style = {
        "streets": {
            "fc": "#2F3737",
            "ec": "#475657",
            "lw": 1,
        }
    },
    radius = False
)

```

```python
# 3. A map showing only small roads and footpaths:

# We'll assume small roads and footpaths to mean residential streets, service roads, footways, cycleways, and pedestrian roads:
plot_small_roads_footpaths = prettymaps.plot(
    'Lansing, Michigan, USA',
    layers = {
        "streets": {
            "tags": {
                "highway": ["residential", "service", "footway", "cycleway", "pedestrian"]
            }
        }
    },
    style = {
        "streets": {
            "fc": "#2F3737",
            "ec": "#475657",
            "lw": 1,
        }
    },
    radius = False
)

```
